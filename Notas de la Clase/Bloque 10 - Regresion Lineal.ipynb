```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 101)
x
```

```python
m = 2
b = 0
y = m * x + b + np.random.normal(0, 1, 101)
```

```python
m_ajuste, b_ajuste = np.polyfit(x, y, 1)
```

```python
plt.scatter(x, y, s = 1.5, c = "black")
plt.plot(x, m_ajuste * x + b_ajuste)
```

```python
ajuste = np.polyfit(x, y, 2)
```

```python
plt.scatter(x, y, s = 0.5)
plt.plot(x, ajuste[0] * x**2 + ajuste[1] * x + ajuste[2])
```
