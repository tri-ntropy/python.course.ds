# Método de Euler

## Importación de bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
import euler
```

## Función de fuerza como una función anónima lambda


```python
fuerza = lambda x, v, i: - x[i - 1] if i <= 5000 else - x[i - 1] - v[i - 1]
fuerza
```

## Inicialización de condiciones iniciales y del tiempo

```python
t0, tf = 0, 50
x0, v0 = 0, 1
n = 12000
```

## Ejecución principal

```python
t, v, x = euler.solver(t0, tf, x0, v0, n, fuerza)
```

## Gráfica

```python
plt.figure(figsize = (15, 4))
plt.plot(t, v, label = "velocidad")
plt.plot(t, x, label = "desplazamiento")
plt.legend(loc = "lower right")
plt.show()
```
