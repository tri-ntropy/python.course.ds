# Graficación

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Estilo de graficación
Seaborn incluye estilos para rapidamente cambiar la apariencia de las gráficas en general

```python
sns.set_style("darkgrid")
```

## Graficación de lineas y dispersión con matplotlib

```python
m = 2
b = 3
x = np.linspace(0, 10, 251)
y = m * x + b + np.random.normal(0, 0.5, 251)
```

Matplotlib ofrece muchas opciones para graficación

```python
plt.plot(x, y, color = "red", ls = ":", label = "$y_1$")
plt.plot(x, y + np.random.normal(2, 0.5, 251), color = "green", ls = "--", label = "$y_2$")
plt.xlabel("$x$")
plt.ylabel("$y(x)$")
plt.legend(loc = "best")
plt.show()
```

```python
plt.scatter(x, y, c = "red", s = 0.5, label = "$y$")
plt.plot(x, m * x + b, c = "black", label = "trayectoria $y$")
plt.xlabel("$x$")
plt.ylabel("$y(x)$")
plt.legend(loc = "best")
plt.show()
```

```python
x = np.linspace(0, 100, 651)
y = np.random.normal(0, 10, 651)
```

```python
plt.figure(figsize = (15, 3))
plt.scatter(x, y, s = 2.0)
plt.xlabel("$t$")
plt.ylabel("$X$")
plt.show()
```

## Graficación de histogramas

```python
plt.hist(y, bins = 25)
plt.show()
```

## Graficación de Heatmaps con matplotlib

```python
X, Y = np.meshgrid(np.linspace(-3, 3, 16), np.linspace(-3, 3, 16))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2) 
```

```python
fig, ax = plt.subplots()

heatmap = ax.imshow(Z)
plt.colorbar(heatmap)

plt.show()
```

## Graficación de heatmaps con seaborn

```python
flights_long = sns.load_dataset("flights")
flights_long
```

```python
flights = (
    flights_long
    .pivot(index="month", columns="year", values="passengers")
)
flights
```

```python
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(flights, annot=False, fmt="d", linewidths=.5, ax=ax, cmap = "gnuplot")
plt.savefig("grafica1.pdf", bbox_inches = "tight")
plt.show()
```
