```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
m = 2
b = 3
x = np.linspace(0, 10, 251)
y = m * x + b
```

```python
plt.plot(x, y, c = "red", ls = "--", label = "$y_1$")
plt.plot(x, y + 2, c = "blue", ls = "-.", label = "$y_2$")
plt.xlabel("$x$")
plt.ylabel("$y(x)$")
plt.hlines(5, 0, 10, colors = "orange", label = "constant")
plt.legend(loc = "best")
plt.savefig("grafica1.pdf", bbox_inches = "tight")
plt.show()
```

```python
plt.scatter(x, y, c = "red", s = 0.5, label = "$y_1$")
plt.scatter(x, y + 2, c = "blue", s = 0.5, label = "$y_2$")
plt.xlabel("$x$")
plt.ylabel("$y(x)$")
plt.hlines(5, 0, 10, colors = "orange", label = "constant")
plt.legend(loc = "best")
plt.show()
```

```python
# make data
X, Y = np.meshgrid(np.linspace(-3, 3, 16), np.linspace(-3, 3, 16))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
```

```python
# plot
fig, ax = plt.subplots()

heatmap = ax.imshow(Z)
plt.colorbar(heatmap)

plt.show()
```

```python
# Load the example flights dataset and convert to long-form
flights_long = sns.load_dataset("flights")
flights_long
```

```python

flights = (
    flights_long
    .pivot(index="month", columns="year", values="passengers")
)
flights
```

```python

# Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(flights, annot=False, fmt="d", linewidths=.5, ax=ax)
```
